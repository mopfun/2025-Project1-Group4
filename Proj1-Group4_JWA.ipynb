```python

import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
#import the data set and create an initial dataframe

anxiety_original_df = pd.DataFrame(pd.read_csv("Resources/anxiety_attack_dataset.csv"))
anxiety_original_df

anxiety_df = anxiety_original_df.rename(columns={'Physical Activity (hrs/week)' : 'Physical Activity', 'Caffeine Intake (mg/day)':'Caffeine Intake', 'Alcohol Consumption (drinks/week)':'Alcohol Consumption', 'Severity of Anxiety Attack (1-10)':'Severity of Anxiety Attack'}, inplace=False)
anxiety_df.sort_values().head(1)
```

##  My Quesitons are:
Do lifestyle factors such as sleep, physical activity, diet, caffeine, smoking, and alcohol intake correlate with the severity of anxiety attacks?
* Sleep Hours
* Physical Activity (hrs/week)
* Caffeine Intake (mg/day)
* Alcohol Consumption (drinks/week)
* Smoking

So let's get the average of each of those for each Severity Level (Severity of Anxiety Attack (1-10))

```python
# Scratch pad

# get the low-down on each contributing factor.
sleep_stats_df = anxiety_df.groupby('Severity of Anxiety Attack')['Sleep Hours'].agg(mean='mean',median='median',max='max',min='min',variance='var',std_dev='std',SEM=lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
physical_stats_df = anxiety_df.groupby('Severity of Anxiety Attack')['Physical Activity'].agg(mean='mean',median='median',max='max',min='min',variance='var',std_dev='std',SEM=lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
caffeine_stats_df = anxiety_df.groupby('Severity of Anxiety Attack')['Caffeine Intake'].agg(mean='mean',median='median',max='max',min='min',variance='var',std_dev='std',SEM=lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
alcohol_stats_df = anxiety_df.groupby('Severity of Anxiety Attack')['Alcohol Consumption'].agg(mean='mean',median='median',max='max',min='min',variance='var',std_dev='std',SEM=lambda x: np.std(x, ddof=1) / np.sqrt(len(x)))
smoking_stats_df = anxiety_df.groupby(['Severity of Anxiety Attack','Smoking'])['Smoking'].count()

caffeine_stats_df.reset_index()
caffeine_stats_df
# print({caffeine_stats_df.loc['min','mean']})

# get teh count of for each of the levels and drinks
# counts = anxiety_df.groupby(['Severity of Anxiety Attack','Caffeine Intake'])['Caffeine Intake'].count()
# counts_df = pd.DataFrame(counts).to_csv("test.csv")
```
